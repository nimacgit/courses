```python
from tqdm import *
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def f(x, l):
    return l*x*(1-x)

def f(x, l):
    return x*x+l

def fix_point(x, l, f):
    for i in range(2000):
        x = f(x, l)
    res = []
    for i in range(500):
        x = f(x, l)
        res.append(x)
    return list(set(res))
        
# l = 3
# x = 0.4
# calcs = []
# calcs.append(x)
# points = [x]
# for i in trange(10000):
#     x = f(x, l)
#     points.append(x)
#     if i % 10 == 0:
#         calcs.append(x)

# plt.figure(figsize=(20, 10))        
# sns.scatterplot(x=range(len(points)), y=points)

res = 500
points = []
for i in trange(res):
    l = 0.25 - 0.001*res
    for x in fix_point(0.321312, l, f):
        points.append([l, x])
points = np.array(points)
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x=points[:, 0], y=points[:, 1])#, s=0.01)
```

```python
from tqdm import *
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 6.9.1, 
# 6.1.8 - 6.1.11 + 6.3.1 - 6.3.7

x = 1
y = 3
t = 0.1
points = []

def yp(x, y):
    return y*y-x*x
def xp(x, y):
    return 2*x*y

for _ in trange(100000):
    points.append([x, y])
    mag = min(abs(yp(x, y)), abs(xp(x, y)))
    t = 0.2**int(np.log10(mag))
#     print(t, mag)
    y1 = y + yp(x, y) * t
    x1 = x + xp(x, y) * t
    x = x1
    y = y1
#     t *= 0.9999
#     print(x, y)
points = np.array(points)
plt.figure(figsize=(20, 10))
sns.scatterplot(x=points[:, 0], y=points[:, 1])
```
