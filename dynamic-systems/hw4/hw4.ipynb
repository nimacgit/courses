```python
from scipy.integrate import solve_ivp
from tqdm import *
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def draw_preiodic(xp, yp, y0, filename):
    def func(t, arg):
        x, y = arg
        return [xp(x,y), yp(x, y)]
    res = solve_ivp(func, [0, 1000], y0)
    xs, ys = res.y
    plt.figure(figsize=(15, 15))
    sns.scatterplot(x=xs, y=ys)
    plt.savefig(f"{filename}.png", dpi=100)

# def xp(x, y):
#     return x-y-x*(x*x+5*y*y)
# def yp(x, y):
#     return x+y-y*(x*x+y*y)

# draw_preiodic(xp, yp, [0.2, 0.1], "inside1")
# draw_preiodic(xp, yp, [5, 4], "outside1")

# def xp(x, y):
#     return x-y-x**3
# def yp(x, y):
#     return y+x-y**3

# draw_preiodic(xp, yp, [0.2, 0.1], "inside3")
# draw_preiodic(xp, yp, [5, 4], "outside3")


def xp(x, y):
    return -x-y+x*(x*x+2*y*y)
def yp(x, y):
    return x-y+y*(x*x+2*y*y)

draw_preiodic(xp, yp, [0.1, 0.1], "inside5")
draw_preiodic(xp, yp, [0.7, 0.7], "outside5")


# dxy = []
# for x,y in zip(xs, ys):
#     dxy.append([xp(x,y)/100, yp(x,y)/100])
# dxy = np.array(dxy)
# dx = dxy[:, 0]
# dy = dxy[:, 1]
# plt.figure(figsize=(30, 30))
# plt.quiver(xs, ys, dx, dy, width=0.0005, headlength=2, headwidth=2, scale=5)
# plt.savefig(file_name, dpi=200)
```
